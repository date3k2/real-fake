```python
import os
import shutil
import fastdup

# Đường dẫn đến thư mục chứa dữ liệu của bạn
data_dir = './drive/MyDrive/Data ML'
# Thư mục đích để chứa các ảnh Real và Fake đã lọc
output_dir = './drive/MyDrive/Processed Data'

# Tạo thư mục Real và Fake trong thư mục đích
# os.makedirs(os.path.join(output_dir, 'Real'), exist_ok=True)
# os.makedirs(os.path.join(output_dir, 'Fake'), exist_ok=True)

def filter_and_move_images(input_dir, category):
    # Sử dụng fastdup để tìm các ảnh trùng lặp
    fd = fastdup.create(input_dir=input_dir, work_dir="AI/")
    fd.run(high_accuracy=True, overwrite=True, nearest_neighbors_k=5)
    duplicate_imgs = set(fd.similarity().sort_values("distance")['filename_from'])

    # Lấy danh sách tất cả các ảnh trong thư mục
    all_images = set(os.listdir(input_dir))

    # Lọc ra các ảnh không trùng lặp
    unique_images = all_images - duplicate_imgs

    # Di chuyển các ảnh không trùng lặp vào thư mục Real/Fake tương ứng
    for img in unique_images:
        if img.lower().endswith(('.jpeg', '.jpg', '.png')):
            src_path = os.path.join(input_dir, img)
            dest_path = os.path.join(output_dir, category, img)
            shutil.copy(src_path, dest_path)

# Duyệt qua tất cả các thư mục con của các thành viên
for folderName, subfolders, filenames in os.walk(data_dir):
    for subfolder in subfolders:
        if subfolder in ["Real", "Fake"]:
            path = os.path.join(folderName, subfolder)
            filter_and_move_images(path,subfolder)
```
