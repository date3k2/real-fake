```python
from diffusers import DiffusionPipeline
import torch

pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, use_safetensors=True, variant="fp16")
pipe.to("cuda")

# if using torch < 2.0
# pipe.enable_xformers_memory_efficient_attention()

prompts = [
    "Create a realistic drawing of a gray tabby cat.",
    "Generate an image of a fluffy white Persian cat.",
    "Draw a picture of a playful orange kitten.",
    "Create an image of a black cat with bright green eyes.",
    "Paint a picture of a Siamese cat lounging in the sun.",
    "Generate an image of a calico cat sitting on a windowsill.",
    "Draw a picture of a sleepy tuxedo cat curled up on a couch.",
    "Create an image of a Maine Coon cat with a majestic mane.",
    "Paint a picture of a ginger cat playing with a ball of yarn.",
    "Generate an image of a cute kitten with blue eyes.",
    "Draw a picture of a fluffy gray cat with a bushy tail.",
    "Create an image of a black and white cat peeking out from behind a plant.",
    "Paint a picture of a Scottish Fold cat with folded ears.",
    "Generate an image of a tabby cat sitting upright and staring curiously.",
    "Draw a picture of a long-haired Ragdoll cat in a cozy bed.",
    # Thêm các prompt khác nếu cần
]
```

```python
for prompt in prompts:
    for i in range(10):
        # Generate image
        image = pipe(prompt=prompt).images[0]
        # Save image with appropriate filename
        image.save(f"./drive/MyDrive/Machine-Learning/{prompt}base-1.0{i}.png")
```
